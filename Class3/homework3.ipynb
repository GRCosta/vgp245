```python
import math
class Circle:

    def __init__(self, radius, x, y):
        """Initialization of the circle"""
        self.x = x
        self.y = y
        self.radius = radius

    def circumference(self):
        """Circumference Calculation"""
        return 2 * 3.14159 * self.radius

    def area(sef):
        """Area of the circle"""
        return 3.14159 * self.radius * self.radius
    
    def location(self): 
        """Locating the circles in space"""
        return "Circle at coordinates (%d, %d)" % (self.x, self.y)


class CircleCollider:

    def __init__(self, x, y, radius):
        """Observed circle for collision"""
        self.x = x
        self.y = y
        self.radius = radius
        

    def collides_with_circle(self, other_circle_collider):
        """Return true or false for the collision check"""
        hipp = math.sqrt((self.x - other_circle_collider.x)**2 + (self.y - other_circle_collider.y)**2)
        return self.radius + other_circle_collider.radius >= (hipp)
    
    
circle1 = Circle(2, 0, 0)
circle2 = Circle(3, 3, 4)

circle1.location()
circle2.location()

collider1 = CircleCollider(circle1.x, circle1.y, circle1.radius)
collider2 = CircleCollider(circle1.x, circle1.y, circle2.radius)

if(collider1.collides_with_circle(collider2)):
    print("Circles Colliding!")
else:
    print("Circles are far apart!")
```
